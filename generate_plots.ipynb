```python
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.utils import plot_model
from tensorflow.keras.models import load_model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score
```

```python
# load models for comparison
# model_mobnetv2 = load_model('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\mobilenetv2\\.mdl_wts_mobilenetv2.hdf5')
model_mobnetv2 = load_model('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\mobilenetv2\\.mdl_wts_mobilenetv2.hdf5')

# plot model using keras plot_model
plot_model(model_mobnetv2, to_file='model_mobnetv2_3classes.png', show_shapes=True, show_layer_names=True)
```

```python
# model_squeezenet = load_model('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\squeezenet\\.mdl_wts_squeezenet.hdf5')
model_squeezenet = load_model('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\squeezenet\\.mdl_wts_squeezenet.hdf5')

# plot model using keras plot_model
plot_model(model_squeezenet, to_file='model_squeezenet_3classes.png', show_shapes=True, show_layer_names=True)
```

```python
# read the history of the models
# for mobilenetv2, history_mobnetv2 is at C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\mobilenetv2\\history_v1.pkl
# for squeezenet, history_squeezenet is at C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\squeezenet\\history_v1.pkl

import pickle
with open('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\mobilenetv2\\history_v1.pkl', 'rb') as file:
    history_mobnetv2 = pickle.load(file)
    
with open('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\model_data\\squeezenet\\history_v1.pkl', 'rb') as file:
    history_squeezenet = pickle.load(file)
```

```python
# let's compare the training and validation history of the two models using seaborn lineplot
# mobilenetv2
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
plt.plot(history_mobnetv2['accuracy'], label='Training Accuracy MobileNetV2', color='blue', linestyle='-')
plt.plot(history_mobnetv2['val_accuracy'], label='Validation Accuracy MobileNetV2', color='red', linestyle='-.')
plt.plot(history_squeezenet['accuracy'], label='Training Accuracy SqueezeNet', color='green', linestyle=':')
plt.plot(history_squeezenet['val_accuracy'], label='Validation Accuracy SqueezeNet', color='orange', linestyle='--')
# plt.title('Training and Validation Accuracy of MobileNetV2')
plt.xlabel('Epochs', fontdict={'size': 11, 'weight': 'bold'})
plt.ylabel('Accuracy', fontdict={'size': 11, 'weight': 'bold'})
plt.xticks(fontsize=11, fontweight="bold")
plt.yticks(fontsize=11, fontweight="bold")
plt.legend(title='Accuracy', loc='center right', title_fontsize=11, fancybox=True, prop={'size': 11, 'weight': 'bold'})
plt.tight_layout()
plt.savefig('C:\\Users\\ishan\\PycharmProjects\\pythonProject1\\data\\Final_Figs\\Accuracy_MobileNetV2_SqueezeNet_c3.png', dpi=400)
plt.show()
```
